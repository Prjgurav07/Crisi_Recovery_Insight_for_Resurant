#### Handling Missing Value

```python
orders.isnull().sum()
```

```python
orders['delivery_partner_id'] = np.where(
    orders['delivery_partner_id'].isnull()
    &
    (orders['is_cancelled']=='Y'),
    "Not Assigned",
    np.where(
        orders['delivery_partner_id'].isnull(),
        "Unknown",
        orders['delivery_partner_id']
    )
)
```

```python
orders.isnull().sum()
```

#### Handling Duplicate data

```python
orders.duplicated().sum()
```

```python
orders.drop_duplicates(subset=['order_id'])
```

#### Covert datetime

```python
orders['order_timestamp'] = pd.to_datetime(orders['order_timestamp'])
```

```python
orders.info()
```

```python
orders['order_year'] = orders['order_timestamp'].dt.year
```

```python
orders['order_month'] = orders['order_timestamp'].dt.month
```

```python
orders['order_day'] = orders['order_timestamp'].dt.day
```

```python
orders['week_day'] = orders['order_timestamp'].dt.day_name()
```

#### Create calculated column

```python
orders['net_revenue'] = orders['subtotal_amount'] - orders['discount_amount'] + orders['delivery_fee']
```

```python
def crisis_phase(date):
    if date < pd.Timestamp('2025-05-31'):
        return "Pre-crisis"
    elif date < pd.Timestamp('2025-06-30'):
        return "Crisis"
    else:
        return "Recovery"
```

```python
orders['crisis_phase'] = orders['order_timestamp'].apply(crisis_phase)
```

```python
delivery = pd.read_csv(r"fact_delivery_performance.csv")
```

```python
delivery.head()
```

```python
delivery.info()
```

```python
def delay_status(row):
    if row['actual_delivery_time_mins'] > row['expected_delivery_time_mins']:
        return 'Delayed'
    else:
        return 'On_Time'
```

```python
delivery['delay_status'] = delivery.apply(delay_status, axis=1)
```

```python
delivery.head()
```

```python
delivery['delay_status'].value_counts()
```

```python
delivery['delivery_delay_mins'] = delivery['actual_delivery_time_mins'] - delivery['expected_delivery_time_mins']
```
